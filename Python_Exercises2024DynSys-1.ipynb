### **1D example: Exponential growth**

We first try to plot the very simple equation $\dot{x} = r x$ which represents exponential growth and its fixed point, which is simply where $\dot{x} = 0 \rightarrow x = 0$.

```python
# Define the equation to plot x' = r x
r = 2;                                                 # Define the value of the parameter
x = np.linspace(-1.0, 1.0, 1000)                       # Define the range of x to plot ([-1,1]) and the number of 
                                                       # points to take (1000)
dxdt = r*x                                             # Define the eq and save it in the variable called dxdt

# Find the value of the Fixed Points and save it in the variable FP
x_sym = sm.symbols('x')                                # Define the symbol x                        
dxdt_Eq = sm.Eq(r*x_sym, 0)                            # Define the symbolic equation rx which must be equal to 0
FP = sm.solve(dxdt_Eq, x_sym)                          # Find the value of the fixed points and save it in FP
```

```python
# Plot
fig,ax = plt.subplots()
show_axes(fig,ax, "x", "x'")
plt.plot(x, dxdt, 'r');                                              # Plot the equation
plt.scatter(FP, np.zeros((np.array(FP).size)))                       # Plot the Fixed Points

# Stylistic adjustments (optional)
plt.legend(['r x', 'Fixed Point'])                                   # Make a legend
plt.title('Exponential growth\n\n', fontsize = 16)                   # Make the title                                                  
plt.grid(True)                                                       # Show the grid in the background

plt.show() 
```

Then we actually solve the differential equation to find the trajectories $x(t)$, using different methods.

```python
# Define the function dxdt to solve
def f_exponential(t,x, r):
    dxdt = r*x
    return dxdt
```

```python
r = 2;                                          # Set the parameters
x0 = [-0.2,0,0.2]                               # Set the initial condition (here we test 3 possible ones)
tboundaries = (0,1)                             # Set the initial time t=0 and the final time of the simulation t = 1
```

### **Built-in function: solve_ivp from scipy.integrate**

```python
# Solve the ODE calling the built-in function odeint from scipy.integrate and find the trajectories xt1, xt2, xt3
sol = solve_ivp(f_exponential, tboundaries, x0, args = (r,), max_step = 0.01)
```

And then we plot it.

```python
plt.rcParams.update({'font.size': 16})
# Plot results
plt.figure(figsize = (13,5))
plt.subplot(121)
for i in range(3):
    plt.plot(sol.t, sol.y[i,:], label = "x0 = "+str(x0[i]))
plt.legend()
plt.xlabel('Time')
plt.ylabel('x(t)')
plt.ylim([-1.5,1.5])
plt.grid(True)

# Plot the vector field using the function quiver
t_vecfield = np.linspace(0, 1,  20)          # Take 20 points between 0 and 1
X_vecfield = np.linspace(-1.5, 1.5, 20)      # Take 20 points between -1.5 and 1.5
T, X = np.meshgrid(t_vecfield, X_vecfield)   # Make a grid of starting points for the arrows in the (t,x) plane
dxdt = r*X                                   # Define the x'(t) for all the X starting points       
plt.subplot(122)
plt.quiver(T, X, np.ones(X.shape), dxdt)     # quiver makes a vector field from the starting points (T,X) with
                                             # arrows components being (1,dxdt)
plt.xlabel('Time'); plt.ylabel('x(t)');
plt.ylim([-1.5,1.5])
plt.show()
```

# Applying the Euler Method

```python
# constants
r = 2.5
x0 = 0.5
t0 = 0
tf = 1
Deltat = 0.1

rho = 25 # controls density of points in the vector field

def derivatives(x):
    return r*x

def euler(x0, t0, tf, Deltat):
    t_arr = np.arange(t0, tf, Deltat)
    x_arr = np.zeros(t_arr.size)
    x_arr[0] = x0
    for i in range(1, t_arr.size):
        x_arr[i] = x_arr[i-1] + Deltat*derivatives(x_arr[i-1])
    return t_arr, x_arr

t_arr, x_arr = euler(x0, t0, tf, Deltat)

t_arr_2, x_arr_2 = euler(-x0, t0, tf, Deltat)

plt.figure(figsize = (13,5))
plt.subplot(121)
plt.plot(t_arr, x_arr)
plt.xlabel('Time')
plt.ylabel('x(t)')
plt.grid(True)

plt.plot(t_arr_2, x_arr_2)


# Plot the vector field using the function quiver
t_vecfield = np.linspace(0, 1,  rho)          # Take 20 points between 0 and 1
X_vecfield = np.linspace(-1.5, 1.5, rho)      # Take 20 points between -1.5 and 1.5
T, X = np.meshgrid(t_vecfield, X_vecfield)   # Make a grid of starting points for the arrows in the (t,x) plane
dxdt = r*X                                   # Define the x'(t) for all the X starting points       
plt.quiver(T, X, np.ones(X.shape), dxdt)     # quiver makes a vector field from the starting points (T,X) with
                                             # arrows components being (1,dxdt)
plt.xlabel('Time'); plt.ylabel('x(t)');
plt.ylim([-1.5,1.5])
plt.show()
plt.close()
```

# Runge-Kutta

```python
# constants
r = 2.5
x0 = 0.5
t0 = 0
tf = 1
Deltat = 0.1

rho = 25 # controls density of points in the vector field

def derivatives(x):
    return r*x

def RK4(x0, t0, tf, Deltat):
    t_arr = np.arange(t0, tf, Deltat)
    x_arr = np.zeros(t_arr.size)
    x_arr[0] = x0
    for i in range(1, t_arr.size):
        k1 = derivatives(x_arr[i-1])*Deltat
        k2 = derivatives(x_arr[i-1] + 1/2 * k1)*Deltat
        k3 = derivatives(x_arr[i-1] + 1/2 * k2)*Deltat
        k4 = derivatives(x_arr[i-1] + k3)*Deltat
        x_arr[i] = x_arr[i-1] + 1/6 * (k1 + 2*k2 + 2*k3 + k4)

    return t_arr, x_arr

t_arr, x_arr = RK4(x0, t0, tf, Deltat)

plt.figure(figsize = (13,5))
plt.subplot(121)
plt.plot(t_arr, x_arr)
plt.xlabel('Time')
plt.ylabel('x(t)')
plt.grid(True)

plt.plot(t_arr_2, x_arr_2)


# Plot the vector field using the function quiver
t_vecfield = np.linspace(0, 1,  rho)          # Take 20 points between 0 and 1
X_vecfield = np.linspace(-1.5, 1.5, rho)      # Take 20 points between -1.5 and 1.5
T, X = np.meshgrid(t_vecfield, X_vecfield)   # Make a grid of starting points for the arrows in the (t,x) plane
dxdt = r*X                                   # Define the x'(t) for all the X starting points       
plt.quiver(T, X, np.ones(X.shape), dxdt)     # quiver makes a vector field from the starting points (T,X) with
                                             # arrows components being (1,dxdt)
plt.xlabel('Time'); plt.ylabel('x(t)');
plt.ylim([-1.5,1.5])
plt.show()
plt.close()
```

> - Investigate the behavoir of the system in the phase space.

```python
# The diff. eq. obeys Liénard's eq. with g(x) = k/m * x

# constants
k = 2
m = 1

x0 = 0
y0 = 0.5
t0 = 0
tf = 10
Deltat = 0.1

def derivatives(x, y):
    v_x = y
    v_y = -k/m * x

    return np.array([v_x, v_y])


def RK4(x0, y0, t0, tf, Deltat):
    t_arr = np.arange(t0, tf, Deltat)
    x_arr = np.zeros(t_arr.size)
    y_arr = np.zeros(t_arr.size)
    x_arr[0] = x0
    y_arr[0] = y0
    for i in range(1, t_arr.size):
        k1_x, k1_y = derivatives(x_arr[i-1], y_arr[i-1])*Deltat
        k2_x, k2_y = derivatives(x_arr[i-1] + 1/2 * k1_x, y_arr[i-1] + 1/2 * k1_y)*Deltat
        k3_x, k3_y = derivatives(x_arr[i-1] + 1/2 * k2_x, y_arr[i-1] + 1/2 * k2_y)*Deltat
        k4_x, k4_y = derivatives(x_arr[i-1] + k3_x, y_arr[i-1] + k3_y)*Deltat
        
        x_arr[i] = x_arr[i-1] + 1/6 * (k1_x + 2*k2_x + 2*k3_x + k4_x)
        y_arr[i] = y_arr[i-1] + 1/6 * (k1_y + 2*k2_y + 2*k3_y + k4_y)

    return t_arr, x_arr, y_arr

t_arr, x_arr, y_arr = RK4(x0, y0, t0, tf, Deltat)

# Plot results in a 3d plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(t_arr, x_arr, y_arr)
ax.set_xlabel('Time')
ax.set_ylabel('x(t)')
ax.set_zlabel('y(t)')
plt.show()
plt.close()

# And in a 2d plot
plt.figure(figsize = (13,5))
plt.subplot(121)
plt.plot(x_arr, y_arr)
plt.xlabel('Time')
plt.ylabel('x(t)')
plt.grid(True)
plt.show()
plt.close()



```

## **Van Der Pol oscillator with limit cycle**

The Van der Pol oscillator is a nonlinear dynamical system that exhibits self-sustained oscillations, meaning it naturally evolves toward a stable periodic motion known as a \textit{limit cycle}. It is governed by the equations:

\begin{align}
    \dot{x} &= y, \\
    \dot{y} &= \mu (1 - x^2) y - x,
\end{align}

where $\mu$ controls the nonlinearity and damping. For small $\mu$, the system behaves like a simple harmonic oscillator, but for larger $\mu$, the oscillations become strongly nonlinear. 

This system has a limit cycle, a characteristically non-linear oscillatory behavior.

# Iterative Maps
We want to investigate another subject that will show up a little later in the course: The iterative map: $x_{n+1} = f(x_n) = rx_n(1-x_n)$, with $0 \leq r \leq 4$, such that the interval $0 \leq x \leq 1$ is mapped onto itself.

> - Try to implement the iterative map and plot the result for different values of r. And different initial conditions. If you just want to play around with it, we've included a solver below.
> - Experiment with the small change in r from r=3.5 to 3=3.6

```python
def logistic_map(x, r):
    return r * x * (1 - x)

def iterate(x, r, n_iterations):
    # Lists with initial values
    xs = [x]
    ys = [0]

    for _ in range(n_iterations):
        # Vertical motion to f(x)
        xs.append(x)
        x = logistic_map(x, r)
        ys.append(x)

        # Horizontal motion: f(x) to y = x
        xs.append(x)
        ys.append(x)

    return xs, ys

def plot(xs, ys, r):
    fig, ax = plt.subplots(figsize=(6,5))

    # Plot polynomial and line y = x
    xs_general = np.linspace(0,1,1000)
    ax.plot(xs_general, logistic_map(xs_general, r), c='k', label='Logistic map')
    ax.plot(xs_general, xs_general, 'k--', label=r'$y = x$')

    # Plot cobweb and markers
    x_color = 'r'
    ax.plot(xs, ys, c=x_color, label='Cobweb')
    ax.scatter(np.zeros_like(xs), xs, c=x_color, marker='_', s=100, alpha=0.1)

    # Format
    ax.set(xlim=(0,1), ylim=(0,1))
    ax.set_xlabel(r'$x_n$', size=16)
    ax.set_ylabel(r'$x_{n+1}$', size=16)
    ax.legend(loc='best')
    fig.tight_layout()

    plt.show()
```

# Implicit methods

A recurring problem when solving differential equations numerically is that of unstable solutions. A good way to get eround that is to use so called **Implicit methods**. The Euler method (which we more precisely may now call the Forward Euler method) is an example of an explicit method, because it uses the value of the function at the current time step to calculate the value at the next time step. The Runge-Kutta method is also an explicit method.

The most simple implicit methods is the **backward Euler method**. It is implicit because it uses the value of the function at the next time step to calculate the value at the current time step.

If $f(x(t), t) = \dot{x}$, we write the explicit Euler method as

$x(t+\Delta t) = x(t) + f(x(t))\Delta t$

and the implicit Euler method as

$x(t+\Delta t) = x(t) + f(x(t+\Delta t), t+\Delta t) \Delta t$

This means that we have to solve an equation to find the value at the current time step. Solving these algebraric equations is usually done with a root finding algorithm, such as the Newton-Raphson method. Since this is not a course on scientific computing, we will not go into general methods for solving equations, but stick to simple examples. If you wish, try experimenting with built-in functions like `fsolve` from the `scipy.optimize` package.

In general, implicit methods might not be totally necessary for the problems we solve here - but they are good to know about, and can be very useful in many cases, especially when solving partial differential equations, where they are often times practically necessary.

# Extra: Speed up your code with Numba

In general, Python is not the fastest language. However, it is very easy to use other languages to speed up your code. One way to do this is to use the `numba` package, which compiles your code to machine code. This is especially useful for loops, which are slow in Python. You may have heard that loops should be avoided in Python, but with `numba` you can often use loops without any problems!

With numba you can use the `@jit` decorator to compile your functions. Usually, `@njit`, which is equivalent to `@jit(nopython = True)` is easiest to use. Notice that the decorator is used to compile *functions*. `numba` only works with a subset of Python, so you may have to change your code a bit to make it work. Generally, it works best with numpy arrays and simple loops and not with library functions from for example `scipy`. For ordinary differential equations, we generally only use numpy arrays and simple loops, so it is often very useful.

```python
from numba import njit, prange
import timeit

# Code with Numba
@njit
def addition_with_numba():
    return 4+4

# Code without Numba
def addition_without_numba():
    return 4+4
    
# Measure execution time with Numba
time_with_numba = timeit.timeit(addition_with_numba, number=1)

# Measure execution time without Numba
time_without_numba = timeit.timeit(addition_without_numba, number=1)



# Print the results
print("Execution time for simple addition without Numba:", time_without_numba)
print("Execution time for simple addition with Numba:", time_with_numba)
```

Above, we actually see that performing a single calculation with `numba` is slower than without. This is because numba has to compile the function before it can be used. However, when we use the function many times, the speedup is significant. The code below is of course not very useful, but it shows how to use `numba` to speed up your code.

```python
import numpy as np
def f(x):
    return -2.3 * x

@njit
def f_numba(x):
    return -2.3 * x



# Define the time step and initial condition
dt = 0.001 # large time step
x0 = 1.0
tend = 10000

# Forward Euler method
def forward_euler():
    t = np.arange(0, tend, dt)
    x = np.zeros(len(t))
    x[0] = x0

    for i in range(1, len(t)):
        x[i] = x[i-1] + f(x[i-1]) * dt

    return x

# Forward Euler method
@njit
def forward_euler_numba():
    t = np.arange(0, tend, dt)
    x = np.zeros(len(t))
    x[0] = x0

    for i in range(1, len(t)):
        x[i] = x[i-1] + f_numba(x[i-1]) * dt

    return x


# Measure execution time with Numba
time_with_numba = timeit.timeit(forward_euler_numba, number=1)

# Measure execution time without Numba
time_without_numba = timeit.timeit(forward_euler, number=1)



# Print the results
print("Execution time for ODE without Numba:", time_without_numba)
print("Execution time for ODE with Numba:", time_with_numba)
```

Way more extreme examples exist, where the speedup is several orders of magnitude. Here an example is shown where 2 lines of code in a loop is sped up 13 times!: https://pythonspeed.com/articles/numba-faster-python/
